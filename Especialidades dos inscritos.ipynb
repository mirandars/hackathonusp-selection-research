# Especialidades

Neste Notebook, criamos algumas visualizações para observar a distribuição de especialidades entre os inscritos.

```python
# Importanto bibliotecas necessárias
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

subs_2016_1 = pd.read_csv('./2016.1/2016.1 - Inscritos.csv', sep=';')
subs_2016_2 = pd.read_csv('./2016.2/2016.2 - Inscritos.csv', sep=';')
subs_2017_1 = pd.read_csv('./2017.1/2017.1 - Inscritos.csv', sep=';')
subs_2017_2 = pd.read_csv('./2017.2/2017.2 - Inscritos.csv', sep=';')
subs_2018_1 = pd.read_csv('./2018.1/2018.1 - Inscritos.csv', sep=';')
```

## 2016.1

```python
subs_2016_1['Qual a sua especialidade?'].value_counts().plot(kind='bar', 
                                          title='Especialidades (2016.1)',
                                          figsize=(10, 10),
                                          label='')
```

```python
freqs = subs_2016_1['Qual a sua especialidade?'].value_counts(normalize=True)
count = subs_2016_1['Qual a sua especialidade?'].value_counts()

for index, value in enumerate(freqs.tolist()):
    print("%s \n Frequência Relativa: %1.2f  Ocorrências: %d\n" %(freqs.index[index], value, count[freqs.index[index]]))
```
