# Institutos / faculdades da USP

Neste Notebook, criamos algumas visualizações para observar a distribuição dos inscritos pelos institutos e faculdades da USP.

```python
# Importanto bibliotecas necessárias
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
subs_2016_1 = pd.read_csv('2016.1-Inscritos.csv', sep=';')
subs_2016_2 = pd.read_csv('2016.2-Inscritos.csv', sep=';')
subs_2017_1 = pd.read_csv('2017.1-Inscritos.csv', sep=';')
subs_2017_2 = pd.read_csv('2017.2-Inscritos.csv', sep=';')
subs_2018_1 = pd.read_csv('2018.1-Inscritos.csv', sep=';')

#df_2019.1 = ger_df() criar função em novo arquivo .py
```

```python
hackathons = [subs_2016_1, subs_2016_2, subs_2017_1, subs_2017_2, subs_2018_1]

for i in hackathons:
    i['Instituto (Sigla)'] = i['Qual a sua escola / faculdade / instituto?'].apply(lambda name: name.split(' ')[-1] if name != 'Não USP' else 'Não USP')
```

## 2016.1

```python
subs_2016_1['Instituto (Sigla)'].value_counts().plot(kind='bar', 
                                          title='Distribuição de inscritos por institutos (2016.1)',
                                          figsize=(20, 10),
                                          fontsize = 14,
                                          #legend = 'True' 
                                          #color = 'Blue'
                                          rot=0
                                                                                                              )
#rodar o texto da base
```
