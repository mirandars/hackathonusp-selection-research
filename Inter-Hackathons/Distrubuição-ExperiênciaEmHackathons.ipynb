# Esperiência em hackathons

Neste Notebook, vamos trabalhar algumas visualizações relativas à distribuição dos inscritos do HackathonUSP quanto ao número de hackathons que o inscrito já participou até o momento de inscrição.

```python
# Importanto bibliotecas necessárias
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

subs_2016_1 = pd.read_csv('./2016.1/2016.1 - Inscritos.csv', sep=';')
subs_2016_2 = pd.read_csv('./2016.2/2016.2 - Inscritos.csv', sep=';')
subs_2017_1 = pd.read_csv('./2017.1/2017.1 - Inscritos.csv', sep=';')
subs_2017_2 = pd.read_csv('./2017.2/2017.2 - Inscritos.csv', sep=';')
subs_2018_1 = pd.read_csv('./2018.1/2018.1 - Inscritos.csv', sep=';')
```

Vale notar que na padronização dos dados, com o objetivo de facilitar a análise, respostas `2 - 4` foram transformadas em apenas `2`. Respotas `Nenhum` foram transformadas em `0`.

## 2016.1

A edição 2016.1 recebeu 550 inscrições, de alunos da USP e também de outras universidades.

```python
subs_2016_1['Você já participou de quantos hackathons?'].value_counts().plot(kind='pie', 
                                          title='Número de hackathons participados (2016.1)',
                                          autopct='%1.1f%%',
                                          figsize=(5, 5),
                                          legend=False,
                                          label='')
```
