# Institutos / faculdades da USP

Neste Notebook, criamos algumas visualizações para observar a distribuição dos inscritos pelos institutos e faculdades da USP.

```python
# Importanto bibliotecas necessárias
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
subs_2016_1 = pd.read_csv('./2016.1/2016.1 - Inscritos.csv', sep=';')
subs_2016_2 = pd.read_csv('./2016.2/2016.2 - Inscritos.csv', sep=';')
subs_2017_1 = pd.read_csv('./2017.1/2017.1 - Inscritos.csv', sep=';')
subs_2017_2 = pd.read_csv('./2017.2/2017.2 - Inscritos.csv', sep=';')
subs_2018_1 = pd.read_csv('./2018.1/2018.1 - Inscritos.csv', sep=';')
```

## 2016.1

```python
subs_2016_1['Qual a sua escola / faculdade / instituto?'].value_counts().plot(kind='bar', 
                                          title='Distribuição de inscritos por institutos/faculdades (2016.1)',
                                          figsize=(20, 10),
                                          )
```
