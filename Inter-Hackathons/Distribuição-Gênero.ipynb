# Gênero

Neste Notebook, vamos trabalhar algumas visualizações relativas à distribuição de gênero dos participantes do HackathonUSP, desde a sua primeira edição (2016.1) até a mais recente (2018.1).

````python
# Importanto bibliotecas necessárias
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
````

Vamos carregar os arquivos .csv com as informações dos inscritos em todas as edições do evento.

````python
subs_2016_1 = pd.read_csv('./2016.1/2016.1 - Inscritos.csv', sep=';')
subs_2016_2 = pd.read_csv('./2016.2/2016.2 - Inscritos.csv', sep=';')
subs_2017_1 = pd.read_csv('./2017.1/2017.1 - Inscritos.csv', sep=';')
subs_2017_2 = pd.read_csv('./2017.2/2017.2 - Inscritos.csv', sep=';')
subs_2018_1 = pd.read_csv('./2018.1/2018.1 - Inscritos.csv', sep=';')
````

As colunas das planilhas estão dispostas da seguinte forma:

```
Carimbo de data/hora --- ID --- Selecionado --- Gênero --- ... --- Time
```

Logo, vamos utilizar a coluna `Gênero` para esta análise.

## 2016.1

Vale notar que essa edição foi _aberta_, ou seja, pessoas que não estudavam na USP poderiam se inscrever.

````python
subs_2016_1['Gênero'].value_counts().plot(kind='pie', 
                                          title='Distribuição de gênero (2016.1)',
                                          autopct='%1.1f%%',
                                          figsize=(5, 5),
                                          fontsize='12',
                                          legend=False,
                                          label='')
````

## 2016.2

````python
subs_2016_2['Gênero'].value_counts().plot(kind='pie', 
                                          title='Distribuição de gênero (2016.2)',
                                          autopct='%1.1f%%',
                                          figsize=(5, 5),
                                          fontsize='12',
                                          legend=False,
                                          label='')
````

## 2017.1

````python
subs_2017_1['Gênero'].value_counts().plot(kind='pie', 
                                          title='Distribuição de gênero (2017.1)',
                                          autopct='%1.1f%%',
                                          figsize=(5, 5),
                                          fontsize='12',
                                          legend=False,
                                          label='')
````

## 2017.2

Esta edição também foi aberta a pessoas que não estudavam na USP.

````python
subs_2017_2['Gênero'].value_counts().plot(kind='pie', 
                                          title='Distribuição de gênero (2017.2)',
                                          autopct='%1.1f%%',
                                          figsize=(5, 5),
                                          fontsize='12',
                                          legend=False,
                                          label='')
````

## 2018.1

````python
subs_2018_1['Gênero'].value_counts().plot(kind='pie', 
                                          title='Distribuição de gênero (2018.1)',
                                          autopct='%1.1f%%',
                                          figsize=(5, 5),
                                          fontsize='12',
                                          legend=False,
                                          label='')
plt.savefig("gender_2018_1.png")
````
